```python
# missing value treatment
```

```python
clean_data['Age'] # here 2,3 has no value we will fill by mean,median,mode for numeric
```

```python
clean_data ['Age']=clean_data['Age'].fillna(np.mean(pd.to_numeric(clean_data['Age'])))
```

```python
clean_data['Age']
```

```python
clean_data['Exp'] # here 3 has no value we will fill by mean,median,mode for numeric
```

```python
clean_data ['Exp']=clean_data['Exp'].fillna(np.mean(pd.to_numeric(clean_data['Exp'])))
```

```python
clean_data['Exp']
```

```python
clean_data
```

```python
# now for catogorical data we use mode/ KNN 
```

```python
clean_data['Location'] # here we have catogorical data or text here 2, 4 hasno value fill by mode /knn
```

```python
clean_data['Location'] =clean_data['Location'].fillna(clean_data['Location'].mode()[0])
```
